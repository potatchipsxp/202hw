```python
%matplotlib inline
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

```python
# 3d Linear system
def ode(state,t):
    x = state[0]
    y = state[1]
    z = state[2]

    # Lorenz with chaotic params
    # sigma = 10
    # rho = 28
    # beta = 8/3 
    
    a = .1
    b = .1
    c = 25

    
    dx = -y - z
    dy = x + a * y
    dz = b + z * (x - c)
    
    # Return the state derivatives
    return [dx,dy, dz]
```

```python
t = np.arange(0, 100, .01) # Times
percent = .4
trunc = int(np.floor(t.size*percent))
fig = plt.figure()
ax = Axes3D(fig)

ic = [4,5,5]
orbit = odeint(ode, ic, t) ### Run the system  
ax.scatter(orbit[0,0],orbit[0,1],orbit[0,2], c="green",s=30)
ax.scatter(orbit[trunc:,0],orbit[trunc:,1],orbit[trunc:,2], s=5)
```
